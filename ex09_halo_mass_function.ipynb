### Inflationary origin of perturbations and their Gaussian pdf

Simplest models of inflation motivate one of the key assumptions in the standard model of structure formation: density perturbations are Gaussian, which means that we can write probability distribution function (pdf) of the density contrast $\delta=\rho/\bar{\rho}-1$ using the usual expression for the Gaussian pdf:

$$
p(\delta)=\frac{1}{\sqrt{2\pi}\sigma}\exp\left[-\frac{\delta^2}{2\sigma^2}\right].
$$

### Variance and rms of the smoothed density field

A related quantity is the variance of the density contrast field 
*smoothed* on some scale $R$ with a window function $W({\bf r}, R)$: 
  
$$\delta_{R}({\bf x})\equiv \int
\delta({\bf x}-{\bf r})W({\bf r}, R)d^3r$$ 
so that variance is 

$$
\langle\delta_{R}^2\rangle\equiv\sigma^2(R)=\frac{1}{(2\pi)^3}\int
P(k)\vert \tilde{W}({\bf k},R)\vert^2d^3k,
$$ 
where $\tilde{W}({\bf k},R)$ is the Fourier transform of the window
(filter) function $W({\bf r},R)$, such that $\delta_{R}({\bf
  k})=\delta({\bf k})\tilde{W}({\bf k},R)$. 
  
At a sufficiently high redshift $z$, for the spherical top--hat window
function mass and radius are interchangeable according to the relation
$M=4\pi/3\rho_{\rm m}(z)R^3$.  We can think about the density field
smoothed on the scale $R$ or the corresponding mass scale $M$. The
characteristic amplitude of peaks in the $\delta_R$ (or $\delta_{M}$)
field smoothed on scale $R$ (or mass scale $M$) is given by
$\sigma(R)\equiv\sigma(M)$.  The smoothed Gaussian density field is,
of course, also Gaussian with the probability distribution function
(PDF) given by 
$$
p(\delta_{M})=\frac{1}{\sqrt{2\pi}\sigma(M)}\exp\left[-\frac{\delta^2_{M}}{2\sigma^2(M)}\right].
$$

### Cold Dark Matter (CDM) to the rescue

After inflation fluctuations of inflaton around its potential minimum are thought to "reheat" the universe and result in a mixture of matter of different kinds and radiation, although details of this process are not yet understood. The inflaton fluctuations are shared by all components: baryons, dark matter, radiation. While their wavelength is above horizon, they just hang around and all wavelenths evolve similarly. 

After a given wavelength becomes smaller than horizon, baryon fluctuations become sound waves (acoustic oscillations) due to large Jeans length imposted by radiation.

Dark matter fluctuations grow, but they do so only logarithmically expansion of the universe is fast, and fluctuations in radiation that dominates energy-density during radiation domination epochs decay. Still, they dark matter fluctuations grow and this logarithmic grows turns out to be sufficient.

### Linear overdensity at collapse

If the initial density contrast $\delta_{\rm i}$ would grow
only at the linear rate, $D_+(z)$, then the density contrast at the time of
collapse would be more than a hundred times smaller.

Its value can be derived starting from the density contrast linearly extrapolated to
the turn around epoch, $\delta_{\rm ta}$. One could expect that at this point the perturbation is already in the *non-linear* regime and so linearly extrapolated overdensity should be of order unity $\delta_{\rm ta}\sim 1$. Because $t_{\rm coll}=2t_{\rm ta}$, further linear
evolution for $\Omega_{\rm m}(z)=1$ until the collapse time gives

$$\delta_c=\delta_{\rm ta} D_+(t_c)/D_+(t_{\rm ta})=a(t_c)/a(t_{\rm ta})=(t_c/t_{\rm ta})^{2/3}=2^{2/3}\approx 1.6$$

In fact, the exact calculation of the evolution equations in the case of 
$\Omega_{\rm m}(z)=1$ at the redshift of turnaround gives $\delta_{ta}=1.062$
(<a href="http://adsabs.harvard.edu/abs/1972ApJ...176....1G">Gunn & Gott 1972</a>), which gives
$\delta_c=\delta_{\rm ta} 2^{2/3}\approx 1.686$.

#### Linearly extrapolated overdensity in $\Omega_{\rm m}\ne 1$ cosmologies

In the
case of $\Omega_{\rm m}\ne 1$ we expect that $\delta_{\rm ta}$ should have
different values. For instance, for $\Omega_{\rm m}< 1$ density contrast at
the turn-around should be higher to account for the higher rate of the
Hubble expansion. However, linear growth from $t_{\rm ta}$ to $t_{\rm
  coll}$ is smaller due to the slower redshift dependence of
$D_+(z)$. As a matter of fact, these two factors nearly cancel, so that 
$\delta_c$ has a weak dependence on
$\Omega_{\rm m}$ and $\Omega_\Lambda$. 

For the concordance
$\Lambda$CDM cosmology at $z=0$, for example, $\delta_c\approx 1.675.$ The exact dependence of $\delta_c$ on cosmological parameters is thus often neglected.

### Mass function of collapsed peaks (aka halos)

Contrasting predictions for the abundance and clustering of collapsed objects
with the observed abundance and clustering of galaxies, groups, and clusters
has been among the most powerful validation tests of structure
formation models (e.g., Press & Schechter 1974, Blumenthal et al 1984, Kaiser 1984, Kaiser 1986).

The abundance of collapsed peaks, called *halos*, is quantified as a function of
their mass, i.e., in terms of the mass function $dn(M,z)/dM$ defined
as the comoving volume number density of halos per the mass interval
$[M,M+dM]$ at a given redshift $z$. This definition is *similar to the luminosity 
function of galaxies* that we've discussed already but instead of luminosity we use mass as label. 

#### Mass function of collapsed halos given a Gaussian field

Statistically, one can define the cumulative probability $F(\geq M,z)$
that a given region within the initial overdensity field smoothed on a
mass scale $M$, $\delta_{\rm M}({\bf x})$, will collapse into a halo
of mass $M$ *or larger* by redshift $z$: 
$$
F(\geq M,z)=\int^{\infty}_{-1}p(\delta_M)C_{\rm coll}(\delta_M,z) d\delta_M,
$$
where $p(\delta_M)d\delta_M$ is the PDF of $\delta_{\rm M}({\bf x})$ for the Gaussian initial density field smoothed on mass scale $M$, and $C_{\rm coll}$ is the *collapse threshold* defining whether any given point ${\bf x}$ with local overdensity $\delta_{\rm M}({\bf x})$ will actually collapse by a given redshift $z$. 

The mass function can then be derived 
as a fraction of the total volume collapsing into halos of mass $(M,M+dM)$, i.e., $dF/dM$, divided by the comoving volume within the initial density field occupied by each such halo, i.e., $M/\bar{\rho}_{\rm}$:
$$
\frac{dn(M)}{dM}=\frac{\bar{\rho}_{\rm m}}{M}\left\vert\frac{dF}{dM}\right\vert.
$$

### Press-Schechter ansatz

In their pioneering model, <a href="http://adsabs.harvard.edu/abs/1974ApJ...187..425P">Press & Schechter (1974)</a> have adopted the
ansatz motivated by the spherical top-hat collapse model  that any point in space with $\delta_{\rm M}({\bf x})D_{+0}(z)\geq \delta_c\approx 1.69\approx\mathrm{const}$ will collapse into a halo of mass $\geq M$ by redshift $z$: i.e., 

$$C_{\rm coll}(\delta)=\Theta(\delta-\delta_c),$$ 

where $\Theta$ is the Heaviside step function. Note that $\delta_{\rm M}({\bf x})$ used above is not the actual initial overdensity, but the initial overdensity evolved to $z$ with the linear growth rate. 

One can easily check that for a Gaussian initial density field this assumption gives 

$$F_{\rm PS}(\geq M,z)=\int^{\infty}_{\delta_c}p(\delta_M) d\delta_M=\frac{1}{2}{\rm erfc}\left[\frac{\delta_c}{\sqrt{2}\sigma(M,z)}\right]=F(\nu).$$ 

where erfc is the <a href="http://mathworld.wolfram.com/Erfc.html">complementary error function</a>: 

$$\mathrm{erfc}(x) = \frac{2}{\sqrt{\pi}}\,\int\limits_x^\infty e^{-t^2}dt.$$

This line of arguments and assumptions thus leads to an important conclusion that *the abundance of halos of mass $M$ at redshift $z$ is a universal function of only their peak height* $\nu(M,z)\equiv\delta_c/\sigma(M,z).$



### Peak height and nonlinear mass

*"Peak height"* is defined as 

$$\nu(M,z)\equiv\frac{\delta_c}{\sigma(M,z)},$$

where $\sigma(M,z)=\sigma(M,z=0)D_+(z).$

According to spherical tophat collapse model, a given overdensity overdensity in a field smoothed on scale $M$ collapses when it reaches linearly extrapolated overdensity of $\delta_c$, while PDF of such overdensities is Gaussian. Peak height, $\nu$, thus measures how rare such collapsing peaks are in the density field. 

For example, $\nu=3$ means that only $\geq 3\sigma$ Gaussian peaks are undergoing collapse and we know that these are quite rare. On the other hand, when $\nu=1$, a typical $1\sigma$ Gaussian fluctuation can collapse. 

Mass scale for which $\nu=1$ at a given $z$ is called *nonlinear mass" and is usually denoted as $M_{\rm NL}$ or as $M_\star$.

### Mass function of collapsed objects in the Press-Schechter ansatz


The fraction of mass in halos per logarithmic interval of mass in the Press-Schechter ansatz is: 

$$
\frac{dn(M)}{d\ln M}=\frac{\bar{\rho}_{\rm m}}{M}\left\vert\frac{dF}{d\ln M}\right\vert=\frac{\bar{\rho}_{\rm m}}{M}\left\vert\frac{d\ln\nu}{d\ln M}\frac{\partial F}{\partial\ln\nu}\right\vert\equiv\frac{\bar{\rho}_{\rm m}}{M}\left\vert\frac{d\ln\nu}{d\ln M}\right\vert\,g(\nu)\equiv \frac{\bar{\rho}_{\rm m}}{M}\,\psi(\nu).
$$

This formula is generic in such approaches and $\psi(\nu)$ depends on assumptions about collapse threshold. For the Press-Schechter ansatz  the mass function of halos has the form

$$\frac{dn_{\rm PS}}{d\ln M}=\sqrt{\frac{2}{\pi}}\frac{\bar{\rho}}{M}\nu\frac{d\ln\nu}{d\ln M} \exp\left(-\frac{\nu^2}{2}\right)$$


#### Press-Schechter mass function

It is instructive to consider prediction of this ansatz for the self-similar model with $\Omega_{\rm m}=1$ and power law power spectrum, $P(k)\propto k^n$ or 

$$\sigma^2(R)\sim k^3 P(k)\vert_{k\sim 1/R}\propto R^{-(3+n)}.$$ 

Fluctuations in the matter-dominated regime in $\Omega_{\rm m}=1$ cosmology grow as $\delta\propto a(t)$ and thus 

$$\sigma\propto a(t) R^{-(3+n)/2}\propto a(t)M^{-(3+n)/6}$$ or $$\nu=\frac{\delta_c}{\sigma}\propto a^{-1}M^{(3+n)/6},$$ 
where $M\propto (1+\delta)R^3$ is mass of the perturbation, and $\delta$ can be neglected in the linear regime $\delta\ll 1$.  

### Power spectrum slope

Recall how the power spectrum slope depends on wavenumber $k=2\pi/\lambda$, where $\lambda=2R$ is perturbation wavelength. For example $R\approx 1.5h^{-1}$ Mpc for $M\approx 10^{12}h^{-1}\ M_\odot$ object, so $k\approx 2$. 

<img src="http://astro.uchicago.edu/~andrey/classes/a304/fig/pk_slope.png" align="center" width="500" height="500"></img>

#### Press-Schechter mass function
So

$$\frac{dn_{\rm PS}}{d\ln M}=\sqrt{\frac{2}{\pi}}\frac{\bar{\rho}}{M}\nu\frac{d\ln\nu}{d\ln M} \exp\left(-\frac{\nu^2}{2}\right);\ \ \ \ 
\nu=\frac{\delta_c}{\sigma}\propto a^{-1}M^{(3+n)/6},$$ 

Thus, for $\nu\ll 1$, $\exp(-\nu^2/2)\approx 1$ and 
$$\frac{dn}{d\ln M}\propto \frac{3+n}{6a} M^{\frac{3+n}{6}-1}.$$ 

At small masses $n$ is close to $n\approx -2.5\div -2.7$ and thus $dn/d\ln M\propto M^{1-\alpha_h}$ with $\alpha_h\approx 1.9\div1.95$. 
Note that $dn/dM = M^{-1}\,dn/d\ln M\propto M^{-\alpha_h}$.

For $\nu\gg 1$

$$\frac{dn}{d\ln M}\propto \exp\left(-\frac{\nu^2}{2}\right)\propto\exp\left(-\frac{AM^{(3+n)/3}}{2}\right)$$

At the scales of the cutoff $n\approx -2.\div -2.5$ typically, so the power of $M$ under exponent is $\approx 0.15\div 0.33$ and the exponential cutoff is quite "tempered" -- i.e. occurs much slower than the cutoff in the galaxy luminosity function. 

### Press-Schechter mass function

Recall that $\nu\propto 1/\sigma(M)\propto M^{-\alpha}$, where $\alpha$ depends on the slope of the power spectrum which changes with scale (and so does $\alpha$). The above form translates into the following mass function as a function of halo mass $M$: approximately a power law at $M\ll M_\star$ and exponential but very tempered cutoff at $M\gg M_\star$.

### "Virial" theorem and the concept of "virialization" 
*see S 4.8 in Binney and Tremain's book "Galactic Dynamics" for details

If one considers a collisionless system, its density distribution in phase space of positions and velocities is governed by the continuity equation, which mathematically expresses the conservation of phase-space volume due to the Liouville's theorem. This theorem states that phase space density is conserved along trajectories in the phase space followed by each particle in the system. This continuity equation is called the <a href="https://ocw.mit.edu/courses/physics/8-902-astrophysics-ii-fall-2004/lecture-notes/lec9.pdf">collisionless Boltzmann equation.</a>  

Integrating the Boltzmann equation over velocities, gives the so-called Jeans equations. While integrating Jeans equations over positions, gives the *tensor virial theorem.* Taking the trace of the tensor virial theorem relation, gives the *scalar virial theorem*, which states that for system in equilibrium (which specifically means that the 2nd derivative of its inertia tensor is zero):

$$2K + W = 0, $$

where 

$$K = \mathrm{trace}\left[K_{jk}\right];\ \ \ K_{jk}=\frac{1}{2}\int \rho\,\overline{v_j v_k}\, d^3{\bf x}$$

$$W=\mathrm{trace}\left[W_{jk}\right];\ \ \ W_{jk}=-\int\rho\, x_k\,\frac{\partial\Phi}{\partial x_j}\, d^3{\bf x}.$$


### The concept of virialization in the tophat collapse model

Tophat collapse solution for $R(t)$ and $\delta(t)$ is formally applicable only until $R>0$. At some point, as perturbation collapses and $R$ approaches $0$, it is assumed that it "virializes" and settles into a spherical distribution with the final radius $R_{\rm f}$ at $t=t_{\rm coll}$. 

Physically, $R_{\rm f}$ is set by the virial relation: 
$2K=-W$, and the total energy $E=K+W=W/2$. At the turnaround $K=0$ as velocity is zero, but total energy is the same, so $E=W_{\rm ta}$. 

Thus, 

$$\frac{GM^2}{R_{\rm ta}}=\frac{GM^2}{2R_{\rm f}}\ \ \mathrm{or}\ \ R_{\rm f}=\frac{R_{\rm ta}}{2}.$$ 

This is not true in cosmologies with $\Omega_{\Lambda}\ne 0$ because in this case vacuum energy is acting as an anti-gravity force and initial energy is not conserved.  
 

### "Virial" density contrast of collapsed peak in the tophat collapse model

The final mean internal density of a collapsed object in a $\Omega_{\rm m}=1$ Universe (with and $\Omega_{\Lambda}=0$) can be estimated
by noting that the time interval $t_{\rm
  coll}-t_{\rm ta}=t_{\rm ta}$ should be equal to the free-fall time
of a uniform sphere $t_{\rm ff}=\sqrt{3\pi/(32G\rho_{\rm ta})}$, which
means that the mean density of perturbation at turnaround is
$\rho_{\rm ta}=3\pi/(32Gt^2_{\rm ta})$ and $\rho_{\rm coll}=8\rho_{\rm
  ta}=3\pi/(Gt_{\rm coll}^2)$. 
  
 These densities can be compared with
background mean matter densities at the corresponding times to get
mean internal density contrasts: $\Delta=\rho/\bar{\rho}_{\rm m}$. 

In the $\Omega_{\rm m}=1$ ($\Omega_{\Lambda}=0$) universe, background density
evolves as $\bar{\rho}_{\rm m}=\bar{\rho}_{\rm m0}a^{-3}$; for $\Omega_{\rm m}=1$ $\bar{\rho}_{\rm m0}=3H_0^2/(8\pi G)$, while $a=(t/t_{\rm U})^{2/3}$, where $t_{\rm U}=2/(3H_0)$ is the current ($a=1$) age of universe. 

These equations combine to give $\bar{\rho}_{\rm m}=1/(6\pi Gt^2)$, which means that density contrast
after virialization is
$$
\Delta_{\rm vir}\equiv \frac{\rho_{\rm coll}}{\bar{\rho}_{\rm m}}=18\pi^2=177.653.
$$

For general cosmologies, density contrast can be computed by
estimating $\rho_{\rm coll}$ and $\bar{\rho}_m(t_{\rm coll})$ in a
similar fashion. For lower $\Omega_{\rm m}$ models, fluctuation of the same mass
$M$ and $\delta$ has a larger initial radius and smaller physical
density and, thus, takes longer to collapse. The density contrasts of
collapsed objects therefore are larger in lower density models because
the mean density of matter at the time of collapse is
smaller. Accurate (to $<1\%$ for $\Omega_{\rm m}=0.1-1$) approximations
for $\Delta_{\rm vir}$ in open ($\Omega_\Lambda=0$) and flat $\Lambda$CDM
($1-\Omega_\Lambda-\Omega_{\rm m}=0$) cosmologies are given by eq. 6 in <a href="http://adsabs.harvard.edu/abs/1998ApJ...495...80B">Bryan & Norman 1998.</a> 

For example, for the concordance $\Lambda$CDM
cosmology with $\Omega_{\rm m}=0.3$ and $\Omega_\Lambda=0.7$,
density contrast at $z=0$ is $\Delta_{\rm vir}\approx 358$.

However, there are serious conceptual issues with using such density contrasts as real physical boundaries of collapsed peaks, because 1) real peaks do not collapse at one time and 2) do not virialize instantly.

### Traixiality of density peaks in the Gaussian field

Elliptical peaks are alongated along major axis and take longer to collapse due to stronger torques from surrounding halos. This means that lower-$\nu$, more elliptical halos have higher density threshold to collapse by a given epoch $z$ than the more spherical high-$\nu$ peaks. 

Triaxiality of peaks can be characterized by their dimensionless principal axes, $\lambda_1\geq\lambda_2\geq\lambda_3$, defined such that $\lambda_1+\lambda_2+\lambda_3=\delta_M$, where $\delta_M$ is overdensity of a peak in the field smooted on scale $M$. <a href="https://link-springer-com.proxy.uchicago.edu/article/10.1007%2FBF01001625">Doroshkevich (1970)</a> showed that for a Gaussian field the joint pdf of $\lambda_1$, $\lambda_2$, $\lambda_3$ is given by 

$$p(\lambda_1,\lambda_2,\lambda_3)=\frac{15^3}{8\pi\sqrt{5}\,\sigma^6}\,\exp\left(-\frac{3\delta^2}{\sigma^2}+\frac{15 I}{2\sigma^2}\right)\,(\lambda_1-\lambda_2)(\lambda_2-\lambda_3)(\lambda_1-\lambda_3), $$

where $\sigma^2=\sigma^2(M)$ is the usual variance of the Gaussian field smoothed on scale $M$, and $I\equiv\lambda_1\lambda_2+\lambda_2\lambda_3+\lambda_1\lambda_3$. So probability of spherical collapse, i.e. $\lambda_1=\lambda_2=\lambda_3$, is actually zero! Collapse is always at least somewhat non-spherical. 

### Ellipsoidal collapse model and corresponding collapse threshold

<a href="http://adsabs.harvard.edu/abs/1996ApJS..103....1B">Bond & Myers (1996)</a> carried out a systematic study of collapse of homogeneous ellipsoids of a given overdensity $\delta$ and trixial shape characterized by the *ellipticity*, $e=(\lambda_1-\lambda_3)/(2\delta)$, and *prolaticity*, $p=(\lambda_1-2\lambda_2+\lambda_3)/(2\delta)$. 

<a href="http://adsabs.harvard.edu/abs/2001MNRAS.323....1S">Sheth, Mo & Tormen (2001,</a> hereafter SMT01) used results of such ellipsoidal collapse model and pdf of $e$ and $p$ for a Gaussian field based on Doroshkevich (1970) to show that linear overdensity at collapse time for homogeneous ellipsoid, $\delta_{\rm ec}$, is given by 

$$\delta_{\rm ec}(e,p)=\delta_{\rm sc}\,\left[1+\beta\left(5(e^2\pm p^2)\frac{\delta^2_{\rm ec}(e,p)}{\delta^2_{\rm sc}}\right)^\gamma\right],$$
where $\delta_{\rm sc}\approx 1.69$ is linear overdensity at collapse time predicted by spherical model. 

This expression can be used for individual peaks with known $e$ and $p$, but to develop a statistical threshold for collapse of ellipsoidal peaks, SMT01 used $p=0$, which corresponds to the peak of the prolaticity pdf, and the value of $e_{\rm mp}=(\sigma/\delta)\sqrt{5}$ that follows a maximum ridge of pdf $g(e,p=0\vert\delta)$. In this case the linear overdensity at collapse becomes:

$$\delta_{\rm ec}=\delta_{\rm sc}\left[1+\beta\left(\frac{\sigma^2}{\delta_{\rm sc}}\right)^\gamma\right]=\delta_{\rm sc}\left[1+\beta\nu^{-\gamma}\right].$$

The parameters $\beta=0.412$ and $\gamma=0.618$ chosen to provide best results (<a href="http://adsabs.harvard.edu/abs/2008MNRAS.388..638D">Desjacques 2008</a>). 

### Linear collapse threshold in the ellipsoidal collapse model

### Ellipsoidal collapse model and corresponding mass function

The ellipsoidal collapse threshold and the Press-Schechter formalism described before, i.e. 

$$
F(M,a)=\int^{\infty}_{-1}p(\delta)C_{\rm ell,coll}(\delta,a) d\delta, \ \ \ C_{\rm coll}(\delta)=\Theta(\delta-\delta_{\rm ec}), 
$$
and
$$
\frac{dn(M)}{dM}=\frac{\bar{\rho}_{\rm m}}{M}\left\vert\frac{dF}{dM}\right\vert.
$$

leads to the following mass function form (<a href="http://adsabs.harvard.edu/abs/2001MNRAS.323....1S">Sheth, Mo & Tormen 2001</a>): 

$$\frac{dn_{\rm SMT}}{d\ln
M}=A\sqrt{\frac{2}{\pi}}\frac{\bar{\rho}}{M}\left(1+\frac{1}{(a\nu)^{2q}}\right)a\nu\frac{d\ln\nu}{d\ln M} \exp\left(-\frac{(a\nu)^2}{2}\right),$$

with $A=0.3222$, $a=0.707$, and $q=0.3.$ The constants and the form of the collapse threshold are adjusted somewhat using calibrations from cosmological simulations of structure formation. For this reason, this function is a pretty accurate  representation of the mass function of collapsed halos in cosmological simulations. It is accurate to better than $30\%$ over galaxy scale masses and has larger error at large masses.

### Code computing halo mass function at a range of $z$ in the ellipsoidal collapse model

```python
import numpy as np
from matplotlib import pylab as plt
from colossus.cosmology import cosmology
import matplotlib.cm as cm
import matplotlib.colors as colors

# define a vector of cosmological parameters:    
my_cosmo = {'flat': True, 'H0': 70.0, 'Om0': 0.27, 'Ob0': 0.045714, 'sigma8': 0.82, 'ns': 0.96}
# set my_cosmo to be the current cosmology	
cosmo = cosmology.setCosmology('my_cosmo', my_cosmo)
#cosmo.interpolation = False

z0=0.0
# define array of redshifts for which to compute and plot mass functions
z = np.array([0.0, 1.0, 2.0, 3.0, 5.0, 7.0, 10.0])

fig = plt.figure(figsize = (3,3))

# define grid of log10 of halo masses in /h Msun
lmh = np.arange(6., 15.5, 0.05)
mh = np.power(10.,lmh) # the actual mass
# compute Lagrangian radius, R=(3M/(4*pi*rho_mean0))^{1/3} 
# corresponding to perturbation of mass mh in /h Mpc
rh = peaks.lagrangianR(mh)
# compute rms fluctuation sigma(R) using default Eisenstein & Hu 1998 power spectrum and tophat filter
sig = cosmo.sigma(rh,z0)
# and its logarithmic derivative dlog(sigma)/dlogR
dsdr = cosmo.sigma(rh, z0, derivative='True')
# dln(sigma)/dlnR -> dln(sigma)/dlnM, M = (4*pi/3) rho_mean * R^3
dsdm = dsdr / 3.0

#
#  SMT01 mass function parameters that were calibrated against numerical simulations
#
q_smt = 0.3; A_smt = 0.322; a_smt = 0.707

# peak height corresponding to mass scale mh, nu=delta_c/sigma(M,z)~1.69/sigma(M,z)
nu = peaks.peakHeight(mh, 0.0)
nup = np.sqrt(a_smt)*nu

# mass function, dn/dlnM, predicted by the Press-Schechter ansatz
# mh is expected in units of h^{-1}, [nps] = [h^3 Mpc^{-3}]
# rho_crit=3H_0^2/(8*pi*G) = 2.77e11 h^2 Msun/Mpc^3
nps = np.sqrt(2.0/np.pi)*cosmo.Om0*2.77e11/mh*nu*np.abs(dsdm)*np.exp(-0.5*nu**2)

# mass function, dn/dlnM, predicted by the SMT01 ansatz based on ellipsoidal collapse model
# mh is expected in units of h^{-1}, [nps] = [h^3 Mpc^{-3}]
nsmt = (A_smt*np.sqrt(2.0/np.pi)*cosmo.Om0*2.77e11/mh * 
        (1.0+1./nup**(2.*q_smt)) * nup * np.abs(dsdm) * np.exp(-0.5*nup**2))

fig = plt.figure(figsize = (3,3))
plt.xlabel(r'$M\ (h^{-1} M_{\odot})$'); plt.ylabel(r'$dn/d\ln M\ (h^3\ \rm Mpc^{-3})$')
plt.yscale('log'); plt.xscale('log')
plt.xlim(1.e6, 4.e15); plt.ylim(1.e-7,1.e3)

cnorm  = colors.Normalize(vmin=0, vmax=22)
scalarmap = cm.ScalarMappable(norm=cnorm, cmap=cm.Blues)

z = np.array([0.0, 1.0, 2.0, 3.0, 5.0, 7.0, 10.0])
lstyles = ['-','--','-.',':','-','--','-.']
nc = np.zeros_like(z)
for i, zd in enumerate(z): nc[i] = np.int(30.*(20.0-zd)/30.0)
color = scalarmap.to_rgba(nc)

for i,zd in enumerate(z):
    nu = peaks.peakHeight(mh, zd)
    nup = np.sqrt(a_smt)*nu
    dsdmz = dsdm*cosmo.growthFactor(zd)
    nsmt = (A_smt*np.sqrt(2.0/np.pi)*cosmo.Om0*2.77e11/mh * 
            (1.0+1./nup**(2.*q_smt))*nup*np.abs(dsdm)*np.exp(-0.5*nup**2))

    plt.plot(mh,nsmt,c=color[i], ls=lstyles[i], linewidth=2.0, label=r'$z=%.1f$'%zd)
    
plt.legend(frameon=False, loc='lower left', fontsize=9)
#plt.grid()
plt.savefig('fig/mfz.png', bbox_inches='tight')
plt.show()
```

### Code comparing SMT01 mass function to Tinker et al. fit to simulations

<a href="http://adsabs.harvard.edu/abs/2008ApJ...688..709T">Tinker et al. (2008)</a> provided a fitting function to halo mass functions measured in cosmological simulations of structure formation accurate to better than $5\%$. The code below compares the SMT01 mass function with a form from the ellipsoidal collapse model to the Tinker et al. 2008 fitting function. 

```python
from colossus.lss import mass_function


z = np.array([0.0, 1.0, 2.0, 3.0, 5.0])
lstyles = ['-','--','-.',':','-','--']
nc = np.zeros_like(z)
for i, zd in enumerate(z): nc[i] = np.int(30.*(20.0-zd)/30.0)
color = scalarmap.to_rgba(nc)

nps = np.zeros((np.size(mh),np.size(z)))
nsmt = np.zeros((np.size(mh),np.size(z)))
diff_ps = np.zeros((np.size(mh),np.size(z)))
diff_smt = np.zeros((np.size(mh),np.size(z)))

for i,zd in enumerate(z):
    nu = peaks.peakHeight(mh, zd)
    nup = np.sqrt(a_smt)*nu
    dsdmz = dsdm*cosmo.growthFactor(zd)
    # mass function, dn/dlnM, predicted by the Press-Schechter ansatz
    # mh is expected in units of h^{-1}, [nps] = [h^3 Mpc^{-3}]
    # rho_crit=3H_0^2/(8*pi*G) = 2.77e11 h^2 Msun/Mpc^3
    nps[:,i] = np.sqrt(2.0/np.pi)*cosmo.Om0*2.77e11/mh*nu*np.abs(dsdm)*np.exp(-0.5*nu**2)
    # SMT01 mass function
    nsmt[:,i] = (A_smt*np.sqrt(2.0/np.pi)*cosmo.Om0*2.77e11/mh * 
            (1.0+1./nup**(2.*q_smt))*nup*np.abs(dsdm)*np.exp(-0.5*nup**2))
    # Tinker et al. 2008 mass function
    mfunc = mass_function.massFunction(mh, zd, mdef = '200m', model = 'tinker08', 
                                       q_out = 'dndlnM')
    # fractional differences
    diff_smt[:,i] = nsmt[:,i]/mfunc - 1.0
    diff_ps[:,i] = nps[:,i]/mfunc - 1.0
```

```python
fig = plt.figure(figsize = (3,3))
plt.xlabel(r'$M\ (h^{-1} M_{\odot})$'); 
#plt.ylabel(r'$\mathrm{SMT01/T08}-1$')
plt.ylabel(r'$n_{\rm PS}/n_{\rm T08}-1$')
plt.xscale('log')
#plt.xlim(1.e8, 4.e15); plt.ylim(-0.3, 0.5)
plt.xlim(1.e9, 4.e15); plt.ylim(-1, 1)

cnorm  = colors.Normalize(vmin=0, vmax=22)
scalarmap = cm.ScalarMappable(norm=cnorm, cmap=cm.Blues)

for i,zd in enumerate(z):

    plt.plot(mh,diff_ps[:,i],c=color[i], ls=lstyles[i], linewidth=2.0, label=r'$z=%.1f$'%zd)
    
plt.legend(frameon=False, loc='lower left', ncol=2, fontsize=7)
#plt.grid()
#plt.savefig('fig/mf_smt01_vs_t08.png', bbox_inches='tight')
plt.savefig('fig/mf_ps_vs_t08.png', bbox_inches='tight')
plt.show()
```
