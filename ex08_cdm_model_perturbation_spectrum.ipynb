### Predictions of the linear evolution model and observations

We see that observed fluctuation amplitude on $10h^{-1}$ Mpc scales between $z_{\rm ls}=1090$ and $z=0$ is $\sim 10^5$, while solution to the linear evolution equation predicts evolution by a factor of $\leq a_{\rm now}/a_{\rm ls}=1+z_{\rm ls}=1090$. 

What is the origin of this discrepancy?

### Jeans instability and Jeans scale

Jeans instability is instability that can operate in self-gravitating, pressurized gas. It is named after James Jeans, who analyzed it and presented criteria for instability in 1902 (<a href="http://adsabs.harvard.edu/abs/1902RSPTA.199....1J">Jeans 1902</a>). When pressure is zero, we don't talk about Jeans instability, we talk anout gravitational instability.

Rigorous treatment of Jeans instability, with dispersion relations etc., can be found in S 5.2.3 of Binney & Tremain's book "Galactic Dynamics" (p. 403). Here I will give a simpler heuristic, but basically correct argument and estimate of the Jeans length scale. 

When pressure forces have a chance to equilibrate against action of gravity, sound crossing time should be *at least* of order dynamical time of collapse of a given region or shorter. To have any chance at being able to "build-up" pressure gradients in response to increased gravity due to a density pertrubation we must have:

$$\frac{\lambda_J}{c_s}\sim t_{\rm dyn}\approx\frac{1}{\sqrt{G\rho}};\ \ \ \ \ c_s\equiv \left(\frac{dP}{d\rho}\right)_S^{1/2},$$

where $c_s$ is speed of sound waves in the medium. Thus,

$$\lambda_J\sim c_s t_{\rm dyn}\approx\frac{c_s}{\sqrt{G\rho}}.$$

Rigorous stability analysis gives:

$$\lambda_J=\sqrt{\pi} c_s t_{\rm dyn}=c_s\sqrt{\frac{\pi}{G\rho}}.$$

Waves with wavelength $\lambda>\lambda_J$ will collapse because pressure gradients cannot be built fast enough. Waves with $\lambda<\lambda_J$ are stable and will simply exist as sound (acoustic) waves.

### Applying Jeans analysis to the baryon plasma at $z_{\rm ls}$

At $z=0$, $\Omega_{\rm b,0}\approx 0.045$. Thus, by definition, baryon density at $z=0$ is:

$$\bar{\rho}_{\rm b,0}=\Omega_{\rm b,0}\rho_{\rm cr,0}=0.045\times 1.88\times 10^{-29} h^2\ {\rm g\,cm^{-3}}\approx 8.46\times 10^{-31} h^2\ {\rm g\,cm^{-3}}$$

Or, assuming that baryons are ionized (i.e., $\rho=\mu_p m_p n_{\rm bar}$ with $\mu_p=0.6$) and $h=0.7$, their number density is: 

$$n_{\rm b,0}=\frac{\rho_{\rm b,0}}{\mu_p m_p}\approx 4.13\times 10^{-7}\,{\rm cm^{-3}}.$$

Thus at $z_{\rm ls}\approx 1090$, 

$$n_{\rm b,z_{ls}}\approx  4.13\times 10^{-7}\, (1+z_{\rm ls})^3\approx 536\,{\rm cm^{-3}}.$$

If the gas shares CMB radiation temperature at that time, it should have temperature:

$$T_{z_{ls}}\approx 2.725\, (1+z_{\rm ls})\approx 2973\,{\rm K}.$$

### Applying Jeans analysis to the baryon plasma at $z_{\rm ls}$

We can numerically write Jeans length for ideal ionized gas as 

$$\lambda_J=c_s\sqrt{\frac{\pi}{G\rho}}=\left(\gamma\,\pi\, \frac{kT}{(\mu_p m_p)^2 G n_{\rm bar}}\right)^{1/2}\approx 1931\,{\rm pc}\,\left(\frac{T}{3300\,\rm K}\right)^{1/2} \, \left(\frac{n_{\rm bar}}{1\,\rm cm^{-3}}\right)^{-1/2},$$

where the last part assumes ideal ionized gas with $\gamma=5/3$. Plugging in $T_{z_{ls}}\approx  2973\,{\rm K}$ and $n_{\rm b,z_{ls}}\approx 536\,{\rm cm^{-3}}$, we get $\lambda_J\approx 79$ pc in physical units, which in comoving scale will be 
$79(1+z_{\rm ls})\approx 8.6\times 10^4$ pc, or 86 kpc. This looks like a galaxy scale, but how much mass such radius would contain?



### "Lagrangian" scale of a galaxy

$$M =\frac{4\pi}{3}(1+\delta)\bar{\rho}_{\rm b,0}\left(\frac{\lambda_J}{2}\right)^3\approx 1.6\times 10^7\ M_\odot$$

Where I assumed that $\delta\ll 1\approx 0$, as is appropriate for early stages. Radius or mass computed with assumption $\delta=0$ are called the *Lagrangian mass and radius* of an object. This mass is quite small. Compare it for example to mass of stars and gas in the Milky Way of $M_{\rm b, MW}\approx 7\times 10^{10}\ \rm M_\odot$. It means that to collect enough mass to form Milky Way, we need to consider a much larger perturbation. Namely: 

$$R=\left(\frac{3M}{4\pi(1+\delta)\bar{\rho}_{\rm b,0}}\right)^{1/3}\approx 1.4\,{\rm Mpc}\,\left(\frac{M_{\rm b}}{7\times 10^{10}\, M_\odot}\right)^{1/3}.$$

This is, of course, much larger than $\lambda_J\approx 86$ kpc and thus such scales should have been Jeans unstable at recombination. 

### Summarizing

* Jeans length on scales from which galaxies like Milky Way collect their mass should have been Jeans unstable at $z_{\rm ls}$. Larger scales are thus also unstable. 
    
    
* However, the amplitude of perturbations is two orders of magnitude smaller than needed to explain existence of $\delta\geq 1$ perturbations we observe on scales $\leq 10$ Mpc at $z=0$. 
    
    
* Why the amplitude of baryon density fluctuations is small at $z_{\rm ls}$?
       
       
* How can this amplitude be reconciled with amplitude of corresponding perturbations at $z=0$?

### Jeans analysis applied to the pre-recombination plasma

We can apply Jeans criterion to the plasma just becore recombination epoch $z\approx z_{\rm ls}\approx 1090$, but we should recall that plasma consists of normal "baryonic" matter and radiation (that includes all relativistic components - in practice, photons and neutrinos):

$$\rho=\rho_\gamma+\rho_{\rm bar}$$

At $z=0$, $\Omega_{\rm r,0}=\Omega_{\gamma 0}+\Omega_{\rm nu 0}\approx 8\times 10^{-5}$, while $\Omega_{\rm b,0}\approx 0.045$. Thus, at $z=0$ baryon density is much higher than radiation density. 

Their actual densities, by definition, are:

$$\bar{\rho}_{\rm b,0}=\Omega_{\rm b,0}\rho_{\rm cr,0}=0.045\times 1.88\times 10^{-29} h^2\ {\rm g\,cm^{-3}}\approx 8.46\times 10^{-31} h^2\ {\rm g\,cm^{-3}}$$

$$\bar{\rho}_{\rm r,0}=\Omega_{\rm r,0}\rho_{\rm cr,0}=8\times 10^{-5}\times 1.88\times 10^{-29} h^2\ {\rm g\,cm^{-3}}\approx 1.5\times 10^{-33} h^2\ {\rm g\,cm^{-3}}$$

where

$$\rho_{\rm cr,0}=\frac{3H_0^2}{8\pi G}\approx 1.88\times 10^{-29} h^2\ {\rm g\,cm^{-3}}.$$

However, the baryon density was higher at a redshift $z$ by $(1+z)^3$, while radiation by $(1+z)^4$ because energy of photons was decreasing with expansion, not just their number density. 

Thus, at $z_{\rm ls}=1090$:

$$\frac{\bar{\rho}_{\rm r}}{\bar{\rho}_{\rm b}}= \frac{\bar{\rho}_{\rm r,0}}{\bar{\rho}_{\rm b,0}}\,(1+z_{\rm ls})\approx 1.8\times 10^{-3}\times 1091\approx 1.94,$$

i.e., radiation energy density dominates and *we have to take radiation into account while computing $c_s$.* For such mix of radiation and baryons the sound speed is (as shown below): 

$$c_s=\frac{c}{\sqrt{3}}\,\left[1+\frac{3}{4}\,\frac{\bar{\rho}_{\rm b}(z)}{\bar{\rho}_{\rm r}(z)}\right]^{-1/2}.$$

### Sound speed of the baryon-radiation plasma $\rho = \rho_\gamma + \rho_{\rm b}$

Pressure of radiation at $z_{\rm ls}$, $P_\gamma=\rho_\gamma c^3/3$, completely dominates over thermal pressure of baryons $P_{\rm b}=nkT$ (by some 9 orders of magnitude), so that

$$c_s=\left(\frac{\partial P}{\partial\rho}\right)^{1/2}_{S=\rm const}=\left(\frac{\partial P_\gamma}{\partial\rho_\gamma}\,\frac{\partial\rho_\gamma}{\partial\rho}\right)^{1/2}_S=\left(\frac{c^2}{3}\left[1+\frac{\partial\rho_{\rm b}}{\partial\rho_\gamma}\right]^{-1}\right)^{1/2}_S=\frac{c}{\sqrt{3}}\,\left[1+\frac{\partial\rho_{\rm b}}{\partial\rho_\gamma}\right]^{-1/2}$$

Recalling that $\rho_\gamma=\rho_{\gamma,0}a^{-4}$ and $\rho_{\rm b}=\rho_{\rm b,0}a^{-3}$, 

$$\frac{\partial\rho_{\rm b}}{\partial\rho_\gamma}=\frac{\partial\rho_{\rm b}}{\partial a}\frac{\partial a}{\partial\rho_\gamma}=\frac{3}{4}\, \frac{\rho_{\rm b,0} a^{-4}}{\rho_{\gamma,0}a^{-5}}=\frac{3}{4}\, \frac{\rho_{\rm b,0} a^{-3}}{\rho_{\gamma,0}a^{-4}}=\frac{3}{4}\, \frac{\rho_{\rm b}(a)}{\rho_{\gamma}(a)}.$$

So

$$c_s=\frac{c}{\sqrt{3}}\,\left[1+\frac{3}{4}\, \frac{\rho_{\rm b}(a)}{\rho_{\gamma}(a)}\right]^{-1/2}$$

At $z_{\rm ls}$ this correspons to $c_s\approx 0.36c$

If baryons were the only matter component in the universe, as we discussed last time, at high $z$, $\Omega_{\rm b}\approx 1$, so $\bar{\rho}=1/(6\pi Gt^2)$ and

$$\lambda_J=c_s\sqrt{\frac{\pi}{G\bar{\rho}}}\approx 0.36c\,\sqrt{\frac{\pi}{G\bar{\rho}}}=0.9\pi ct.$$

At the same time the horizon scale of the universe at this time can be computed as follows. Recall (ex01, 1st class) that when photons redshift by $\delta z$ they travel comoving distance $\delta d=c/H(a)\delta z=c/(\dot{a}/a)\delta z=cdt\,a(\delta z/da)=cdt/a$, where I used $dz/da=-1/a^2$. Thus, when the age of the universe is $t$ they travelled the comoving distance (where we use $a(t)=(t/t_U)^{2/3}$ appropriate for $\Omega_{\rm b}\approx 1$):

$$d_h = \int\limits_0^t\frac{cdt}{a}=\frac{3ct}{a}$$

and thus, the proper horizon scale will be $\lambda_h=a\times 3ct/a=3ct$.

Thus, $\lambda_J>\lambda_h$ and *all scales smaller than horizon are Jeans stable.* All baryon fluctuations evolved as sound waves, with approximately constant amplitude. Thus, **the reason baryon fluctuations are so small at $z_{\rm ls}$ is that they could not grow before that epoch due to large Jeans length set by radiation.**

After recombination, the radiation and photons decouple and Jeans length of baryons is set by baryon sound speed. This decreases baryon Jeans length by a factor of $\approx 2.6\times 10^{-5}$ to $\approx 80$ pc with the corresponding Jeans mass of $\approx 10^7\ M_\odot$.


### The fluctuation amplitude crisis

This is why once upper limits on the fluctuations of the CMB temperature became smaller than $\delta T/T\sim 10^{-3}$ circa 1980, the observed structures in the universe could no longer be explained under standard gravity filled with only baryonic matter. 

The reconciliation within standard general relativity proposed in a series of papers in 1982 is that in addition with baryonic matter our universe is filled with weakly interacting particles, commonly known as **Cold Dark Matter (CDM)**. 

Dark matter is not coupled to photons (it does not have EM interaction, and hence is dark) and is thus does not care about Jeans length set by photons. Weak self-interactions of dark matter particles implies that dark matter "soup" (for the lack of better word, because "plasma" is reserved for systems mediated by EM interactions) has no pressure to counteract gravity. Thus, perturbations in dark matter grow via gravitational instability, even when baryon fluctuations do not grow due to large Jeans length. 

### Inflationary origin of perturbations and their Gaussian pdf

Simplest models of inflation motivate one of the key assumptions in the standard model of structure formation: density perturbations are Gaussian, which means that we can write probability distribution function (pdf) of the density contrast $\delta=\rho/\bar{\rho}-1$ using the usual expression for the Gaussian pdf:

$$
p(\delta)=\frac{1}{\sqrt{2\pi}\sigma}\exp\left[-\frac{\delta^2}{2\sigma^2}\right].
$$

where $\sigma^2$ is the variance of the $\delta$ field. 

### Variance and rms of the smoothed density field

A related quantity is the variance of the density contrast field 
*smoothed* on some scale $R$ with a window function $W({\bf r}, R)$: 
  
$$\delta_{R}({\bf x})\equiv \int
\delta({\bf x}-{\bf r})W({\bf r}, R)d^3r$$ 
so that variance is 

$$
\langle\delta_{R}^2\rangle\equiv\sigma^2(R)=\frac{1}{(2\pi)^3}\int
P(k)\vert \tilde{W}({\bf k},R)\vert^2d^3k,
$$ 
where $\tilde{W}({\bf k},R)$ is the Fourier transform of the window
(filter) function $W({\bf r},R)$, such that $\delta_{R}({\bf
  k})=\delta({\bf k})\tilde{W}({\bf k},R)$. 
  
At a sufficiently high redshift $z$, for the spherical top--hat window
function mass and radius are interchangeable according to the relation
$M=4\pi/3\rho_{\rm m}(z)R^3$.  We can think about the density field
smoothed on the scale $R$ or the corresponding mass scale $M$. The
characteristic amplitude of peaks in the $\delta_R$ (or $\delta_{M}$)
field smoothed on scale $R$ (or mass scale $M$) is given by
$\sigma(R)\equiv\sigma(M)$.  The smoothed Gaussian density field is,
of course, also Gaussian with the probability distribution function
(PDF) given by 
$$
p(\delta_{M})=\frac{1}{\sqrt{2\pi}\sigma(M)}\exp\left[-\frac{\delta^2_{M}}{2\sigma^2(M)}\right].
$$

### Power spectrum of fluctuation amplitudes

The total variance of the unsmoothed density field $\langle\delta^2\rangle$ is often quantified in terms of the *power spectrum* defined as:

$$
\langle\delta^2\rangle=\frac{1}{(2\pi)^3}\int d^3\vec{k}P(k)=\frac{1}{2\pi^2}\int P(k)k^2 dk 
= \frac{1}{2\pi^2}\int P(k)k^3 d\ln k = \int\Delta^2(k)d\ln k, 
$$

Statistical properties of a uniform and isotropic Gaussian field depend
only on the modulus $k$ of the wavevector, but not on its direction. Hence, vectors disappear in the above expressions. 

In the expression above $\Delta^2(k)$ is dimensionless power per logarithmic decade of $k$:
$$\Delta^2(k)=\frac{4\pi k^3}{(2\pi)^3}P(k)=\frac{1}{2\pi^2}k^3 P(k).$$

Note that $\langle\delta^2\rangle$ is dimensionless, so that $k^3P(k)$ is dimensionless, i.e., $P(k)$ has units of volume. 

Overall, the power spectrum $P(k)$ is computed as the average of the square amplitude of the Fourier components of perturbations with wavevector amplitude $k$ (only the amplitude matters because we assume isotropic universe):
$$P(k)\equiv \langle |\delta_k|^2 \rangle $$
where average is computed over some shell $[k,k+dk]$ in the Fourier space, and is normalized by the volume of the shell. See script in this notebook for an example of how this is done.

### Brief history of density fluctuations: inflationary origin

Fast exponential expansion for $\sim 60-70$ e-foldings during inflation epoch corresponds to the increase of scales by $>e^{60-70}\sim 10^{26}-10^{30}$ so atomic scale becomes $\rm pc-kpc$ scale. This is physical scale. The corresponding comoving scale is very large. Thus, even quantum scales get inflated to macroscopic and cosmological scales. According to inflation theory, *all that grand-scale structure in the galaxy distribution, resulted from the quantum fluctuations of the inflaton field.*

A quantum fluctuation in the inflaton energy $\delta E$ can be viewed as a wave packet of wavelength $\lambda\sim hc/\delta E$. The horizon scale is 
$$R_h=a(t)\int\limits_0^t\frac{cdt^\prime}{a(t^\prime)}\sim ct\sim \frac{c}{H(t)}\approx\rm const,$$
 but physical scales increase as $a(t)\propto\exp(Ht)$ and so most scales will become larger than the horizon with the smallest scales leaving the horizon last. 

When the wavelength of perturbation 
becomes equal to horizon $\lambda\sim R_h\propto 1/\delta E\sim\rm const$. In GR, $\delta E\propto \Delta\Phi$ (potential perturbations. Thus, $\delta\Phi\approx\rm const$ at $t$ when $\lambda(t)\approx R_h$. 


This is the physical origin of the name *scale-invariant perturbation power spectrum*. The nearly scale-invariant spectrum of potential fluctuations is a generic prediction of inflation scenario, because the slow-roll period and the associated regime of $H\approx\rm const$ are generic to this scenario, even as details of models vary. 

### scale-invariant power spectrum

Potential fluctuations are related to the density fluctuations via the Poisson equation: $\nabla^2\delta\Phi=4\pi G\bar{\rho}\delta$, where $\delta=(\rho-\bar{\rho})/\bar{\rho}$. 

or in Fourier space: $k^2\delta_{\Phi,k}\propto \delta_{\rho,k}$. Thus 

$$\Delta^2_{\rho}(k)\propto k^4\Delta^2_{\Phi}(k)$$ 

and scale-invariant power spectrum of potential fluctuations $\Delta^2_{\Phi}\approx\rm const$ corresponds to the density power spectrum 

$$\Delta^2(k)\propto k^{4}\ \mathrm{or}\ P(k)\propto k,$$ 

which is also called scale-invariant (which is somewhat confusing) or often the *Harrison-Zeldovich spectrum.* 

The <a href="http://adsabs.harvard.edu/abs/2016A%26A...594A..13P">current constraints from the Planck observatory</a> indicate initial perturbation spectrum very close to the Gaussian (deviations from Gaussianity were not detected at a fairly stringent level) and the power spectrum shape on the largest scales: $P(k)\propto k^n$ with  

$$n=0.968\pm 0.006$$

This level of deviation from $n=1$ is also predicted generically by inflation models due to the generic predictions of existence of scalar and tensor perturbation modes and the fact that horizon is not exactly constant in time. 

### Cold Dark Matter (CDM) to the rescue

After inflation fluctuations of inflaton around its potential minimum are thought to "reheat" the universe and result in a mixture of matter of different kinds and radiation, although details of this process are not yet understood. The inflaton fluctuations are shared by all components: baryons, dark matter, radiation. While their wavelength is above horizon, they just hang around and all wavelenths evolve similarly. 

After a given wavelength becomes smaller than horizon, baryon fluctuations become sound waves (acoustic oscillations) due to large Jeans length imposted by radiation.

Dark matter fluctuations grow, but they do so only logarithmically expansion of the universe is fast, and fluctuations in radiation that dominates energy-density during radiation domination epochs decay. Still, they dark matter fluctuations grow and this logarithmic grows turns out to be sufficient.

```python
from colossus.cosmology import cosmology

# define a vector of cosmological parameters:    
my_cosmo = {'flat': True, 'H0': 70.0, 'Om0': 0.27, 'Ob0': 0.045714, 'sigma8': 0.82, 'ns': 0.96}
# set my_cosmo to be the current cosmology
cosmo = cosmology.setCosmology('my_cosmo', my_cosmo)
cosmo.interpolation = False

# input model spectrum for comparison
lka = np.arange(-4.,3.,0.02); ka = 10.0**lka
Pka = cosmo.matterPowerSpectrum(ka)
cosmo.interpolation = True
dPdka = cosmo.matterPowerSpectrum(ka, derivative=True)

#
# plot input model P(k) 
#
plot_pretty()
plt.figure(figsize=(3,3))
plt.xscale('log'); plt.yscale('log')
plt.ylim(1.e-2, 1.e5); plt.xlim(1.e-4, 1.e2)
plt.plot(ka, Pka, c='slateblue', linewidth=2., 
         label=r'$\Lambda\mathrm{CDM}:\ \Omega_{\rm m}=%.1f\ \Omega_\Lambda=%.1f$'%(cosmo.Om0,cosmo.Ode0))
plt.legend(frameon=False, loc='lower left', fontsize=7, scatterpoints=1)
plt.xlabel(r'$k\ (h\rm\ Mpc^{-1})$')
plt.ylabel(r'$P(k)\ (h^{-3}\ \rm Mpc^3)$')
plt.grid(linestyle='dotted', c='lightgray')
#plt.savefig('fig/pk.pdf', bbox_inches='tight')
plt.show()
# plot Delta^2(k)
plt.figure(figsize=(3,3))
plt.xscale('log'); plt.yscale('log')
plt.xlim(1.e-4, 1.e2); plt.ylim(1.e-9, 1.e2)
plt.plot(ka, Pka*ka**3/(2.*np.pi**2), c='slateblue', linewidth=2., 
         label=r'$\Lambda\mathrm{CDM}:\ \Omega_{\rm m}=%.1f\ \Omega_\Lambda=%.1f$'%(cosmo.Om0,cosmo.Ode0))  
plt.legend(frameon=False, loc='lower right', fontsize=7, scatterpoints=1)
plt.xlabel(r'$k\ (h\rm\ Mpc^{-1})$')
plt.ylabel(r'$\Delta^2(k)$')
plt.grid(linestyle='dotted', c='lightgray')
#plt.savefig('fig/Delta2k.pdf', bbox_inches='tight')
plt.show()

# plot slope of P(k) = dlnP(k)/dlnk = k/P*dP/dk
plt.figure(figsize=(3,3))
plt.xscale('log'); 
plt.xlim(1.e-4, 1.e2); plt.ylim(-3, 1.)
plt.plot(ka, dPdka, c='slateblue', linewidth=2., 
         label=r'$\Lambda\mathrm{CDM}:\ \Omega_{\rm m}=%.1f\ \Omega_\Lambda=%.1f$'%(cosmo.Om0,cosmo.Ode0))  
plt.legend(frameon=False, loc='lower left', fontsize=7, scatterpoints=1)
plt.xlabel(r'$k\ (h\rm\ Mpc^{-1})$')
plt.ylabel(r'$d\ln P(k)/d\ln k$')
plt.grid(linestyle='dotted', c='lightgray')
#plt.savefig('fig/pk_slope.pdf', bbox_inches='tight')
plt.show()
```

```python
import numpy as np
from numpy.fft import fftn, ifftn, fftfreq
from numpy.random import normal
from matplotlib import pylab as plt

from colossus.cosmology import cosmology

# define a vector of cosmological parameters:    
my_cosmo = {'flat': True, 'H0': 70.0, 'Om0': 0.3, 'Ob0': 0.045714, 'sigma8': 0.82, 'ns': 0.96}
# set my_cosmo to be the current cosmology
cosmo = cosmology.setCosmology('my_cosmo', my_cosmo)
cosmo.interpolation = False

# box size in h^{-1} Mpc
Lbox = 100.0
# grid size (change N to 128 if 256 is taking too long)
N=256; N3 = N*N*N
dx = Lbox/N; dxi = 1.0/dx

# set up grid coordinates
x1d = np.linspace(0, N-1, N)
x, y, z = np.meshgrid(x1d, x1d, x1d)

# set random seed; remove if you want new realization every time
np.random.seed(141592653)
# generate 3D array of Gaussian numbers with zero mean and unit dispersion
delta = np.reshape(normal(0.0, 1.0, N3),(N,N,N))

# Fourier transform delta field
delta = fftn(delta)
# get vector of wavenumbers corresponding to modes, fftfreq takes care of mode "packing" order
k1d = fftfreq(N) * 2.0 * np.pi
# grid of k in Fourier space
kx, ky, kz = np.meshgrid(k1d, k1d, k1d)
k = dxi * np.sqrt(kx**2 + ky**2 + kz**2)
k[0,0,0] = np.pi * 2.0 / Lbox

# set up power spectrum for the input model at redshift zini
zini = 30.0
# growth factor at 30. and its square
gzi = cosmo.growthFactor(30.0); gzi2 = gzi * gzi
# normalize the power spectrum to z=30
Pk  = cosmo.matterPowerSpectrum(k) * gzi2 

# normalize Fourier components using sqrt(P(k))
delta *= np.sqrt(Pk)
# this is "DC" or "fundamental mode" corresponding to constant density in the box. 
# it can be set to any constant
delta[0,0,0] = 0.0

# compute and check P(k)
lk = np.log10(k)
Nb = 20 # number of k bins to use
# P(k) itself - is just a weighted histogram of squares of delta_k amplitudes
Pkj,  lkj = np.histogram(lk, bins=Nb, weights=np.square(np.absolute(delta)), density=False) 
# <delta^4> for variance of the P(k)
Pkj2, lkj = np.histogram(lk, bins=Nb, weights=np.square(np.square(np.absolute(delta))), density=False) 
lkaj, lkj = np.histogram(lk, bins=Nb, weights=lk, density=False) 
nj , lkj  = np.histogram(lk, bins=Nb, density=False) 
# P(k) and its rms error due to finite number of modes
Pkj = Pkj / nj / gzi2 / N3
Pkj2 = Pkj2 / (gzi2*gzi2) / N3 / N3
sPkj = np.sqrt(Pkj2/(nj*(nj-1)) - Pkj**2/(nj-1))
lkaj = lkaj / nj 
kaj = 10.**lkaj

# transform the field back into real space
delta = np.real(ifftn(delta))

# input model spectrum for comparison
lka = np.arange(-4.,3.,0.02); ka = 10.0**lka
Pka = cosmo.matterPowerSpectrum(ka)

#
# plot input model P(k) and P(k) for the realization
#
Pkjp = np.clip(Pkj,1.e-10,1.e100)
plt.figure(figsize=(3,3))
plt.xscale('log'); plt.yscale('log')
plt.ylim(1.e-2, 1.e5); plt.xlim(1.e-4, 1.e2)
plt.errorbar(kaj, Pkjp, yerr=2.*sPkj, fmt='o', markersize=1.)
plt.scatter(kaj, np.clip(Pkj,1.e-10, 1.e100), s=10., label=r'$\mathrm{realization}$')
plt.plot(ka, Pka, c='slateblue', linewidth=2., label=r'$\mathrm{input\ model}$')
plt.legend(frameon=False, loc='lower left', fontsize=10, scatterpoints=1)
plt.xlabel(r'$k\ (h\rm\ Mpc^{-1})$')
plt.ylabel(r'$P(k)\ (h^{-3}\ \rm Mpc^3)$')
#plt.grid()
plt.savefig('fig/pk_realization.pdf', bbox_inches='tight')
plt.show()
```

```python
from scipy import signal

def Wfk(kR, filter = 'tophat'):
    """
    Top-hat and Gaussian window functions in the Fourier space
    """
    if filter == 'tophat':
        return 3.*(np.sin(kR)-kR*np.cos(kR))/(kR)**3
    elif filter == 'gaussian':
        return np.exp(-0.5*kR**2)
    else:
        print "error in Wfk: routine accepts only tophat or gaussian filter"
        return

#
# plot a slice of the density field of Gaussian realization to check things visually
#
import matplotlib.ticker as mtick
plot_pretty()
lloc = 'upper right'; lfs = 9; lcol = 'white'; ccol = 'yellow'
fig, ax = plt.subplots(2,2,figsize=(4,4))

for ix in [0,1]:
    for iy in [0,1]:
        ax[ix][iy].set_xlim(0., Lbox); ax[ix][iy].set_ylim(0., Lbox)

dmin = 0.75*np.min(delta[:,:,N/2]); dmax = 0.5*np.max(delta[:,:,N/2])
x1d = Lbox*np.linspace(0, N-1, N)/(N-1)
x, y = np.meshgrid(x1d, x1d)

ax[0][0].imshow(delta[:,:,N/2], cmap='winter', vmin=dmin, vmax=dmax, extent=(0., Lbox, Lbox, 0.))
ax[0][0].plot(0, 0, "-", c='black', label=r'$\mathrm{no\ smoothing}$' )
#ax[0][0].contour(x, y, delta[:,:,N/2]/gzi.T, extent=(0., Lbox, 0., Lbox), linewidths=(0.5), colors=ccol, levels = [1.69])

#ax[0][0].colorbar(fraction=0.04, pad=0.04)
lgnd = ax[0][0].legend(frameon=False, loc=lloc, fontsize=lfs)
for text in lgnd.get_texts():
    text.set_color(lcol);
for item in lgnd.legendHandles:
    item.set_visible(False)
 
# smooth the field with a given filter size and plot
Rfa = [0.5, 1.5, 5.0] #/h Mpc
ix = [0, 1, 1]; iy = [1, 0, 1]
# smooth by a real-space top-hat filter
filter = 'tophat'

for ir, Rf in enumerate(Rfa):
    kR = k*Rf; 
    deltas = fftn(delta) * Wfk(kR, filter)
    deltas = np.real(ifftn(deltas))
    #ax[ix[ir]][iy[ir]].imshow(deltas[:,:,N/2], cmap='winter', vmin=dmin, vmax=dmax, extent=(0., Lbox, 0., Lbox) )
    ax[ix[ir]][iy[ir]].imshow(deltas[:,:,N/2], vmin=dmin, vmax=dmax, 
                              cmap='winter',  extent=(0., Lbox, Lbox, 0.) )
    delta_s = deltas[:,:,N/2]/gzi
    ax[ix[ir]][iy[ir]].contour(x, y, delta_s,  linewidths=(0.5), colors=ccol, 
                               levels = [1.69])
    ax[ix[ir]][iy[ir]].plot(0, 0, "-", c='black', label=r'$R_{\rm f}=%.1fh^{-1}\ \mathrm{Mpc}$'%Rf )
    lgnd = ax[ix[ir]][iy[ir]].legend(frameon=False, loc=lloc, fontsize=lfs)
    for text in lgnd.get_texts():
        text.set_color(lcol);
    for item in lgnd.legendHandles:
        item.set_visible(False)

plt.setp(ax[0][0].get_xticklabels(), visible=False)
plt.setp(ax[0][1].get_xticklabels(), visible=False)
plt.setp(ax[0][1].get_yticklabels(), visible=False)
plt.setp(ax[1][1].get_yticklabels(), visible=False)
ax[0][0].set_ylabel(r'$y\ (h^{-1}\ \mathrm{Mpc})$')
ax[1][0].set_ylabel(r'$y\ (h^{-1}\ \mathrm{Mpc})$')
ax[1][0].set_xlabel(r'$x\ (h^{-1}\ \mathrm{Mpc})$')
ax[1][1].set_xlabel(r'$x\ (h^{-1}\ \mathrm{Mpc})$')

plt.tight_layout()

plt.savefig('fig/gfield_smooth.png', bbox_inches='tight')
plt.show()
```
